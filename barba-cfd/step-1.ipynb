```python
import numpy as np 
import matplotlib.pyplot as plt
import time, sys
```

```python
%matplotlib inline
```

```python
width = 2 # spatial domain of 2 unit leghths wide
```

```python
number_of_grid_points = nx = 81
dx = width / (nx-1) 
print(dx)
number_of_time_step = nt = 25
dt = 0.025 # delta t -> amount of time each timestep covers
wave_speed = c  = 1 # wave speed = 1
```

```python
u = np.ones(nx)
```

```python
print(u)
```

```python
## u_0 = initial_velocity  (initilization ) = 2 in the interval 0.5 <= x <= 1 
```

```python
print(int((width/4)/dx))
print(int((width/2)/dx+1))
u[int(0.5/dx):int(1/dx+1)]=2
```

```python
print(u)
```

```python
plt.plot(np.linspace(0,2,nx),u)
plt.show()
```

```python
## I have observed that when i reduced the number of grid points the line became slant and when i increased the count tro 81, it was becoming more straight without the slantness, this is indicative of the fact that when we make it more fine, the simuls change. 
```

```python
un = np.ones(nx)
print(un)
```

```python
for n in range(number_of_time_step):
    un = u.copy()
    plt.plot(np.linspace(0,2,nx),un)
    plt.show()
    for i in range(1,nx):
        u[i] = un[i]-c*dt/dx*(un[i]-un[i-1])
        
```

```python
plt.plot(np.linspace(0,2,nx),u)
plt.show()
```

```python
## in lower resolutions of grid spacing -- say 41 - we could see that approximations distorted the shape. even at 81, it is distoring the shape, since we are always utilizing the last approximated data to compute the next, the gradient multiplications is desolving the resolution -- this is called numerical diffusion -- using taylor series we can explain and expand this.  
```
